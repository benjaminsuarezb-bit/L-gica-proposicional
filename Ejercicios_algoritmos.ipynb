```python
# Ejercico 12

horas_trabajadas = 48
valor_hora = 5000
porcentaje_retencion = 12.5 / 100  # 12.5%

# El salario bruto es lo que el empleado gana sin descuentos
salario_bruto = horas_trabajadas * valor_hora
# La retención de la fuente es un porcentaje del salario bruto por lo que podemos pasar el porcentaje a numero decimal y multiplicarlo
retencion = salario_bruto * porcentaje_retencion
# El salario neto es lo que realmente recibe el trabajador después del descuento por retención
salario_neto = salario_bruto - retencion

print("Salario bruto: $", salario_bruto)
print("Retención en la fuente: $", retencion)
print("Salario neto: $", salario_neto)
```
